### 2. Filtering data by IMDB score

#### 2.1 

Write a function that:

1. Accepts a single movie dictionary from the `movies` list as an argument.
2. Returns True if the IMDB score is above 5.5.

#### 2.2 [Challenge] 

Write a function that:

1. Accepts the movies list and a specified category.
2. Returns True if the average score of the category is higher than the average score of all movies.

````python
# Your code here.
def check_score(single_mv_dict, score):
    if single_mv_dict["imdb"] > score:
        return True
    else:
        return False
    
print movies[6]
print check_score(movies[6],5.5)

def avg_category_score (category,movie_list):
    if category == 'all':
        return sum([x["imdb"] for x in movie_list])/len([x["imdb"] for x in movie_list])
    else:
        return sum([x["imdb"] for x in movie_list if x["category"].lower() == category.lower()])/len([x["imdb"] for x in movie_list if x["category"].lower() == category.lower()])

        
def check_category(category, movie_list):
    avg_cat_score = avg_category_score(category,movie_list)
    avg_overall_score = avg_category_score('all',movie_list)
    #print avg_cat_score, avg_overall_score
    if avg_cat_score > avg_overall_score:
        return True
    else:
        return False

print check_category('suspense',movies)
print check_category('Crime',movies)
````

### 3. Creating subsets by numeric condition

#### 3.1

Write a function that:

1. Accepts the list of movies and a specified imdb score.
2. Returns the sublist of movies that have a score greater than the specified score.

#### 3.2 [Expert] 

Write a function that:

1. Accepts the movies list as an argument.
2. Returns the movies list sorted first by category and then by movie according to average score and individual score, respectively.

````python
from pprint import pprint
# Your code here.
def subset(score, movie_list):    
    filtered_list = []
    for movie in movie_list:
        if check_score(movie,score) == True:
            filtered_list.append(movie)
    return filtered_list

pprint(subset(2.5, movies))
#pprint(movies)



def sort_movies(movies_list):
    unique_cat = {}
    scores = []
    sorted_cat = []
    sorted_movies = []
    for x in movies_list:
        unique_cat[x['category']] = 1
    for y in unique_cat:
        scores.append((y, avg_category_score(y,movies_list)))
    #pprint(scores)
    sorted_cat = sorted(scores, key = lambda scores : scores[1])
    sorted_cat.reverse()
    print '\nHere are the category average scores'
    pprint(sorted_cat)
    for z in sorted_cat:
        temp = [(a,b['imdb'])  for a,b in enumerate(movies_list) if b['category']==z[0]]
        cat_movies = sorted(temp, key = lambda temp : temp[1])
        cat_movies.reverse()
        for c in cat_movies:
            sorted_movies.append(movies_list[c[0]])
    return sorted_movies       

print '\nThe list has now been sorted below'       
pprint(sort_movies(movies))
    
````

````python
def category_subset(category, movie_list):    
    filtered_list = []
    for movie in movie_list:
        if check_category(movie,category) == True:
            filtered_list.append(movie)
    if len(filtered_list) == 0:
        print category, 'does not exist'
        return None
    else:
        return filtered_list

pprint(category_subset('romance', movies))

def search_movie1(search_string, movie_list):
    temp = list(set([x['category'] for x in movie_list if search_string.lower() in x['category'].lower()]))
    temp2 = list(set([x['name'] for x in movie_list if search_string.lower() in x['name'].lower()])) 
    return {'category':temp, 'name':temp2}
    
pprint(search_movie1('ri',movies))
````
