```python
import scrapy
import pandas as pd
import numpy as np
import time
import requests
import pandas
from scrapy.selector import Selector
from scrapy.http import HtmlResponse
import cPickle as pickle

domain = "https://www.indeed.com"
webpath = "http://www.indeed.com/resumes?q=%22data+scientist%22+OR+%22data+analyst%22+OR+%22business+analyst%22&l=San+Francisco%2C+CA"
webpath = "http://www.indeed.com/resumes?q=%22data+scientist%22+OR+%22data+analyst%22+OR+%22business+analyst%22&l=New+York%2C+NY"
webpath = "http://www.indeed.com/resumes?q=%22data+scientist%22+OR+%22data+analyst%22+OR+%22business+analyst%22&l=San+Jose%2C+CA"



 
label = 'SF'
resume_output = []
for i in range(1,2):
    response = requests.get(webpath)
    HTML = response.text
    links =  Selector(text=HTML).xpath("//ol[@id='results']/li/div/div/div/a/@href").extract()
    idz = Selector(text=HTML).xpath('//li[@class="sre"]/@id').extract()
    # no next links --- all on one page
    next_link = Selector(text=HTML).xpath("//div[@id='pagination']/a[@class='confirm-nav next']/@href").extract()
    for x,y in zip(idz,links):
        print x,y
        subwebpath = domain + y
        print subwebpath        
        response = requests.get(subwebpath)
        if response.status_code =='200':
            subHTML = response.text
            localdict = {
                "url":subwebpath
                ,"label":label
                ,"id":x
                , "html":subHTML
            }
            resume_output.append(localdict)
        else:
            print 
            break
    print next_link
    webpath = domain + next_link[0]
    
print len(resume_output)
file_i = 1


```
