# Stagin the linked in data

```python
with open('master_lnk_df.p','rb') as f:
    master_lnk_df = pickle.load(f)
master_lnk_df.info()
```

```python
master_lnk_df[['head-location','body-location']]
```

```python

all_states = ['AL','AK','AZ','AR','CA','CO','CT','DE','FL',
 'GA','HI','ID','IL','IN','IA','KS','KY','LA',
 'ME','MD','MA','MI','MN','MS','MO','MT','NE',
 'NV','NH','NJ','NM','NY','NC','ND','OH','OK',
 'OR','PA','RI','SC','SD','TN','TX','UT','VT',
 'VA','WA','WV','WI','WY']
```

```python

master_lnk_df.dropna(how='any',inplace=True)
master_lnk_df['city'] = master_lnk_df['head-location']
master_lnk_df['city'] = master_lnk_df['city'].map(lambda x : str(''.join([a for a in x if ord(a)<128])))

def swap(x):
    x = x.replace('US','')
    for y in all_states:
        x = x.replace(y,'')
    x = x.replace(',','')
    return x.strip()

def isolateState(x):
    for y in all_states:
        if y in x:
            return y
        
master_lnk_df['city'] = master_lnk_df['city'].map(swap)
master_lnk_df['sourcesite'] = 'lnk'
master_lnk_df['state'] = master_lnk_df['body-location'].map(isolateState)
#master_lnk_df['state'] = master_lnk_df['body-location'].map(lambda x : x) 
master_lnk_df.info()
master_lnk_df.head(2)
```

```python
master_lnk_df['state'].value_counts()
```

```python
to_merge1 = master_ind_df[['company_x', 'snip', 'city','state','jobtitle','sourcesite','views','days_posted','post_start_date','url']].copy()
to_merge1.columns = ['company','jobdesc','city','state','title','sourcesite','views','days_posted','post_start_date','link' ]

to_merge2 = master_lnk_df[['body-company', 'jobdesc', 'city', 'state', 'head-title','sourcesite','views','days_posted','post_start_date','link' ]].copy()
to_merge2.columns = ['company','jobdesc','city','state','title','sourcesite','views','days_posted','post_start_date','link']
```

```python
master_total_df = pd.concat([to_merge1,to_merge2],axis=0)
master_total_df.title.fillna('',inplace=True)
master_total_df['title'] = master_total_df['title'].map(lambda x : str(''.join([a for a in x if ord(a)<128])))
master_total_df.reset_index(inplace=True)

master_total_df
```

```python
base_titles = ['engineer','analyst','scientist','specialist', 'developer',
               'architect','consultant','manager','programmer','director',
               'accountant','scientist','designer','assistant','intern','lead','vp','vice president',
               'researcher', 'administrator','associate','statistician','editor'
              ]
master_total_df['base_title'] = master_total_df['title'].map(lambda x : ''.join([' ' if a in (';:./,-') else a for a in x]))
master_total_df['base_title'] =  master_total_df['base_title'].map(lambda x :[a for a in x.lower().split(' ') if a in base_titles])
master_total_df['base_title'] = master_total_df['base_title'].map(lambda x : x[0] if len(x)>0 else '')

master_total_df['parsed_title'] = master_total_df['title'].map(lambda x : ''.join([' ' if a in (';:./,-') else a for a in x]))
master_total_df['parsed_title'] = master_total_df['parsed_title'].map(lambda x : x.lower().split(' '))
master_total_df['parsed_title_i'] = master_total_df['parsed_title'].map(lambda x : [i for i, a in enumerate(x) if a in base_titles])

def expandedTitle(x):
    if len(x['parsed_title_i'])==0:
        return np.nan
    elif x['parsed_title_i'][-1] == 0:
        return x['parsed_title'][0].strip()
    else:
        return str(x['parsed_title'][int(x['parsed_title_i'][-1])-1] + ' '+ x['parsed_title'][int(x['parsed_title_i'][-1])]).strip()

def prefixTitle(x):
    if len(x['parsed_title_i'])==0:
        return np.nan
    elif x['parsed_title_i'][-1] == 0:
        return x['parsed_title'][0].strip()
    else:
        return str(x['parsed_title'][int(x['parsed_title_i'][-1])-1]).strip()
 
    
master_total_df['expanded_title'] = master_total_df.apply(expandedTitle,axis=1)
master_total_df['prefix_title'] = master_total_df.apply(prefixTitle,axis=1)
master_total_df.head(3)
```
