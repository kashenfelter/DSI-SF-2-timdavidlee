```python
from os import listdir
import os.path
import pandas as pd
from os.path import isfile, join
import numpy as np
import json
from bs4 import BeautifulSoup
from scrapy.selector import Selector
import datetime

folder_location = '/Users/tlee010/desktop/DSI-capstone/indeed-api-dump/02-indeed_job_details/'
onlyfiles = [f for f in listdir(folder_location) if isfile(join(folder_location,f)) & ('json' in f)]

# start loop

jobkeys =[]
urls = []
titles = []
companys = []
locations = []
jobdescs = []
snips = []
htmls = []
i = 0
filecount = 0
for single_file in onlyfiles: 
    if i % 100==0:
        print i
    filecount +=1
    print single_file, filecount, 'out of', len(onlyfiles) , datetime.datetime.now()
    filepath = folder_location + single_file
    with open(filepath, 'rb') as f:
        tjson = json.loads(f.read())

    for subjob in tjson['job_details']:
        i +=1
        local_html = subjob['HTML'].encode('utf-8')
        title=Selector(text =local_html) .xpath('//b[@class="jobtitle"]/font/text()').extract()
        company=Selector(text =local_html) .xpath('//div[@data-tn-component="jobHeader"]/span[@class="company"]/text()').extract()
        location=Selector(text =local_html) .xpath('//div[@data-tn-component="jobHeader"]/span[@class="location"]/text()').extract()
        jobdesc= Selector(text =local_html) .xpath('//span[@class="summary"]/text()').extract()
        snip = Selector(text=local_html).xpath("//td[@class='snip']").extract()
        html_raw = local_html 
        snipsoup = BeautifulSoup(''.join([x for x in snip]))
        for elem in snipsoup.findAll(['script', 'style']):
            elem.extract()

        try:
            title = title[0].encode('utf-8')
        except:
            print 'error: ',title
            title = np.nan

        
        try:
            company = company[0].encode('utf-8')
        except:
            print 'error: ',company
            company = np.nan

        
        try:
            location = location[0].encode('utf-8')
        except:
            print 'error: ',location
            location = np.nan

        
        try:
        
            jobdesc = '\n'.join([x.strip().encode('utf-8') for x in jobdesc])
        except:
            jobdesc = np.nan
        
        try:
            snipsoup = ''.join([x.encode('utf-8') for x in snipsoup.get_text()])
        except:
            print 'snip exception'
            snipsoup = np.nan

        jobkeys.append(subjob['id'])
        urls.append(subjob['url'])
        titles.append(title)
        companys.append(company)
        locations.append(location)
        jobdescs.append(jobdesc)
        snips.append(snipsoup)
        htmls.append(html_raw)
df = pd.DataFrame({
        "jobkey": jobkeys
        ,"url":urls
        ,"title":titles
        ,"company":companys
        ,"location":locations
        ,"jobdesc":jobdescs
        ,"snip":snips
        ,"html":htmls
    })

print df.shape
df.head()
df.to_csv('job_details.json')
df.to_pickle('job_details.p')
```
