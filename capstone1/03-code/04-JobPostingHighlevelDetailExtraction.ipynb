```python
import numpy as np
from os import listdir
import cPickle as pickle
import requests
import json
import time
from os.path import isfile, join
folder_location = '../indeed-api-dump/01-indeed_links/'
onlyfiles = [f for f in listdir(folder_location) if isfile(join(folder_location,f)) & ('json' in f)]
print onlyfiles

def getJobdesc(master_json, fileprefix, filename):
    accum_list = []
    
    all_jobs = {"job_details":[]}
    i = 0
    for x in master_json['processing results']:
        for y in x['response']['results']['result']:
#             print y
#             print y.keys()
#             print y['formattedLocation']['$']
#             print y['formattedLocationFull']['$']
#             print y['city']
#             print y['jobtitle']['$']
#             print y['source']['$']
#             print y['state']['$']
#             print y['date']['$']
#             print y['jobkey']['$']
#             print y['company']['$']
#             try:
#                 print y['latitude']['$']
#                 print y['longitude']['$']            
#             except:
#                 print 'no lat/long'
            local_dict = {}
            local_dict['formattedLocation'] = y['formattedLocation']['$']
            local_dict['formattedLocationFull'] = y['formattedLocationFull']['$']
            try:
                local_dict['city'] = y['city']['$']
            except:
                local_dict['city'] = y['city']
            local_dict['jobtitle'] = y['jobtitle']['$']
            local_dict['source'] = y['source']['$']
            local_dict['state'] = y['state']['$']
            local_dict['date'] = y['date']['$']
            local_dict['jobkey'] = str(y['jobkey']['$'])
            try:
                local_dict['company'] = y['company']['$']
            except:
                local_dict['company'] = y['company']
            try:
                local_dict['latitude'] = y['latitude']['$']
                local_dict['longitude'] = y['longitude']['$']
            except:
                local_dict['latitude'] = np.nan
                local_dict['longitude'] = np.nan
            local_dict['filename'] = filename
            i +=1
            accum_list.append(local_dict)
    return accum_list

global_list = []
for x in onlyfiles:
    thisjson = json.load(open(folder_location + x,"rb"))
    print '%s is open ' % (x)
    temp = getJobdesc(thisjson,x[:3], x)
    global_list.extend(temp)
    

with open('jobposting_summary_details.p','wb') as f:
    print len(global_list)
    pickle.dump(global_list,f)
    print 'pickling dictionary complete'
    
import pandas as pd
dedup = {
    x['jobkey']:x for x in global_list
}

dedup_list = [x for x in dedup.values()]
df = pd.DataFrame(dedup_list)
with open('jobposting_summary_df.p','wb') as f:
    pickle.dump(df,f)
    print 'pickling dataframe complete'

    
```
