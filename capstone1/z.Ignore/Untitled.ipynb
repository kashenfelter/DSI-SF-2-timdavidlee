```python
import scrapy
import pandas as pd
import numpy as np
import time
import requests
from scrapy.selector import Selector
from scrapy.http import HtmlResponse

domain = 'http://www.simplyhired.com'
webpath = 'http://www.simplyhired.com/search?q=data+analyst%2C+data+scientist%2C+business+analyst&l=san+francisco%2C+ca&mi=50'
response = requests.get(webpath)
HTML = response.text 
ind = 0
master_data = []
for i in range(1,3):
    response = requests.get(webpath)
    HTML = response.text 
    pagelinks =[x.encode('utf-8') for x in Selector(text=HTML).xpath("//a[@class='card-link js-job-link']/@href").extract()]
    nextlink = Selector(text=HTML).xpath("//a[@class='next-pagination']/@href").extract()[0]
    
    for j in pagelinks:
        print i,ind
        newlink = domain+j
        #print newlink
        time.sleep(np.random.randint(1,5))
        response = requests.get(newlink)
        localHTML = response.text
        master_data.append({"index":ind, "link":j, "html_page":localHTML})
        ind +=1

    print domain + Selector(text=HTML).xpath("//a[@class='next-pagination']/@href").extract()[0]   
    webpath = domain + nextlink
```

```python
import pandas
df = pandas.DataFrame({'index':[x['index'] for x in master_data] ,'link':[x['link'] for x in master_data],'html_page':[x['html_page'] for x in master_data] })
```
