```python
#rename columsn
rock_clean1 = rock_data.rename(columns={
        'Song Clean': 'songname'
        , 'ARTIST CLEAN' : 'artistname'
        , 'Release Year' : 'releaseyear'
        , 'COMBINED':'artist_song_key'
        , 'Year?' : 'releaseyear_available'
        , 'PlayCount': 'plays_per_year'
        , 'F*G': 'total_plays'
    })

# clean the column heads
print rock_clean1.head(5)
print rock_clean1['releaseyear'].unique()
rock_clean1['releaseyear'].fillna('None')
rock_clean1['releaseyear'].replace({'SONGFACTS.COM':'None', np.nan:'None'},inplace=True)
print rock_clean1['releaseyear'].unique()
```

```python
#**1. What are the top 20 most popular songs by plays?**
print rock_clean1[['songname','total_plays']].groupby('songname').sum().sort_values('total_plays',ascending = False)[1:20]

#**2. Which years have the most plays?**
print rock_clean1[['releaseyear','total_plays']].groupby('releaseyear').sum().sort_values('total_plays',ascending = False)[1:20]

#**3. Which records don't have matching "Play Count" cooresponding to "F*G"?**
print rock_data[rock_data['F*G']!=rock_data['PlayCount']]
# --- where there is no year value (or no year is available)




```
