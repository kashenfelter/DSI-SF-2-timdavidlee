## 7. Calculate player "power".

---

Define a function that will calculate a player's "power". Player power is defined as the sum of the attack and defense values of all of their pokemon.

Your function will:

1. Accept the `players` dictionary, `pokedex` dictionary, and a player_id as arguments.
2. For the specified player_id, look up that player's pokemon and their level(s).
3. Find and aggregate the attack and defense values for each of the player's pokemon from the `pokedex` dictionary.
4. Print "[player name]'s power is [player power].", where the player power is the sum of the attack and defense values for all of their pokemon.
5. Return the player's power value.

Remember, the attack and defense of a pokemon in the `pokedex` dictionary are lists ordered by level, so you will need to pull out the attack and defense value indexed by the pokemon's level!

Print out the pokemon power for each of your players.

````python
# corrected text:

"""
Calculate player "power".

Define a function that will calculate a player's "power". Player power is defined as the sum of the base stats for a pokemon.

Your function will:
- Accept the players dictionary, pokedex dictionary, and a player_id as arguments.
- For the specified player_id, look up that player's pokemon.
- Find and aggregate the base stats values for each of the player's pokemon from the pokedex dictionary.
- Print "[player name]'s power is [player power].", where the player power is the sum of the attack and defense values for all of their pokemon.
- Return the player's power value.

Print out the pokemon power for each of your players.
"""

# calculate pokemon power individually, adds up selected stats. The stats is a list of metrics to add
def calc_poke_power(poke_dict,pokeid,stats):
    return sum([(poke_dict[pokeid][stat]) for stat in stats])

#calculate total player power = sum of all pokemon
def calc_player_power(player_dict, poke_dict, pid):
    
    #identifies all pokemon for a particular player into a list
    players_poke = [x for x in player_dict[pid]['player_pokemon']]
    
    #sets baseline power
    player_power = 0
    
    #which aspects will be added up to total power
    aspects = ['attack','defense','special_attack','special_defense']
    player_power += sum([calc_poke_power(poke_dict,y,aspects) for y in players_poke])
    
    #it will both printout within the function and then return a number
    print '%s\'s power is %d.' % (player_dict[pid]['player_name'], player_power)
    return player_power

print "\nFor Reference, all current pokemon stats:"
pprint(poke_dict)
pprint([calc_poke_power(poke_dict,x,['attack','defense','special_attack','special_defense']) for x in poke_dict.keys()])

print "\nCharacter power"
for pid in player_dict.keys():
    calc_player_power(player_dict, poke_dict, pid)
    
# Example output:
#
# Ash's power is 57.
# 57
# Brock's power is 119.
# 119
````

````python
# Code to read in pokedex info
raw_pd = ''
pokedex_file = 'pokedex_basic.csv'
with open(pokedex_file, 'r') as f:
    raw_pd = f.read()
    
# the pokedex string is assigned to the raw_pd variable
````

````python
# Example output (first 3 sublists):
#
# ['PokedexNumber', 'Name', 'Type', 'Total', 'HP', 'Attack', 'Defense', 'SpecialAttack', 'SpecialDefense', 'Speed']
# [1.0, 'Bulbasaur', 'GrassPoison', 318.0, 45.0, 49.0, 49.0, 65.0, 65.0, 45.0]
# [2.0, 'Ivysaur', 'GrassPoison', 405.0, 60.0, 62.0, 63.0, 80.0, 80.0, 60.0]
def uptype_data(inlist):
    newlist = []
    for x in inlist:
        if type(x) in (int,float):
            newlist.append(1.0*x)
        elif type(x) == str and x.isdigit():
            newlist.append(float(x))
        elif type(x) == str:
            newlist.append(x)
        elif type(x) == str and len(x) == 0:
            newlist.append(-1)
        else:
            newlist.append(str(x)+" error")
    return newlist

raw_rows = raw_pd.split('\n')
print raw_rows[0]
clean_data = []
#first row
clean_data.append(raw_rows[0].replace('"','').split(','))


for row in raw_rows[1:]:
    clean_data.append(uptype_data(row.replace('"','').split(',')))

print '\nLimit to a few samples'
pprint(clean_data[:6])

````
